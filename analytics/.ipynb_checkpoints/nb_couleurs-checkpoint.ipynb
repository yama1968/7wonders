```python
%matplotlib inline
import scipy.stats as stats
from IPython.core.pylabtools import figsize
import numpy as np
import pandas as pd
figsize(12.5, 8)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
scores = pd.read_csv("../data/Scores 7 Wonders - Scores.csv")
scores["Rang relatif"] = (scores["Rang relatif"].str[:-1].astype(float) / 100)
scores.head()
```

```python
scores.columns
```

```python
couleurs = scores.set_index(["Partie", "Joueur"])[[u'Rouge', u'Argent', u'Merveille', u'Bleu', u'Jaune',
       u'Violet', u'Vert']]
couleurs_r = couleurs.divide(couleurs.sum(axis = 1), axis = 0)
couleurs_r.head()
```

```python
from biokit.viz import corrplot

df = pd.merge(scores[["Nb joueurs", "Rang relatif", "Vainqueur", "Partie", "Joueur"]], 
              couleurs_r.reset_index(),
              on = ["Partie", "Joueur"]).set_index(["Partie", "Joueur"])

df.head()
```

```python
c = corrplot.Corrplot(df.corr())
c.plot(method = "pie")

c1 = corrplot.Corrplot(df[df.Vert > 0.3].corr())
c1.plot(method = "pie")
```
